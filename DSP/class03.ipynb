```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sr = 100
t = np.arange(0, 1, 1/sr)
```

```python
f = 1
s = (np.sin(2*np.pi * f * t) )
```

```python
plt.plot(t, s)
plt.axis([0,1,-1,1])
```

```python
f2 = 2
s2 = np.sin(2*np.pi *f2 *t)
```

```python
plt.plot(t,s)
plt.plot(t, s2)
plt.plot(t, s + s2)
plt.axis([0,1,-2,2])
```
